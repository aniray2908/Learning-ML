```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\Anisha Ray\Downloads\py-master\ML\1_linear_reg\Exercise\canada_per_capita_income.csv")
df
```

```python
df.columns=[c.replace(' ', '_') for c in df.columns]
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per_capita_income_(USD)'])
```

```python
reg.predict([[2020]])
```

```python
m = reg.coef_
b = reg.intercept_
x = 2020
y = m*x + b
y
```

```python
%matplotlib inline
plt.xlabel('year', fontsize=10)
plt.ylabel('Per Capita Income (US$)', fontsize=10)
plt.scatter(df['year'], df['per_capita_income_(USD)'], color='red', marker='*')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```
